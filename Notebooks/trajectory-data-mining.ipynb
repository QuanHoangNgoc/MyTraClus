```python
import os
import matplotlib.pyplot as plt
from traclus.traclus import traclus, smooth_trajectory
import numpy as np 
```

```python
data = []
path = '/kaggle/input/tdriver/release/taxi_log_2008_by_id'
for file in os.listdir(path):
    file_path = os.path.join(path, file)
    sto = []
    with open(file_path, 'r') as f:
        for i, line in enumerate(f):
            content = line.split(',')
            x, y = float(content[-2]), float(content[-1])
            sto.append(np.array([x, y]))
    data.append(np.array(sto))
```

```python
partitions, segments, dist_matrix, clusters, cluster_assignments, representative_trajectories = traclus(data[0:10])
```
